```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
path = './healthcare_noshows.csv'

my_data = pd.read_csv(path)
```

```python
my_data.head()

# my_data.dropna(axis=0, subset=['ScheduledDay', 'AppointmentDay'], inplace=True)

# drop idåˆ—
my_data.drop(['PatientId', 'AppointmentID'], axis=1, inplace=True)
```

```python
my_data.head()
```

```python
plt.figure(figsize=(16,6))

sns.lineplot(data=my_data, x='Age', y='Alcoholism', hue='Gender', style='Gender', markers=True, dashes=False)
```

```python
# KDE
plt.figure(figsize=(16,6))
sns.kdeplot(data=my_data, x='Age', shade=True, hue='Gender', multiple='stack', palette='coolwarm')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```
